```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load movie databases and merge in pandas
movie_file = pd.read_csv('movies.csv')
ratings_file = pd.read_csv('ratings.csv')
df = pd.merge(movie_file, ratings_file)
df
```

```python
# pivot table
ratings = pd.pivot_table(df, index='userId', columns='title', values='rating')
ratings.head(3)
```

```python
ratings = ratings.reindex(ratings.count().sort_values(ascending=False).index, axis=1)
ratings.head(3)
```

```python
ratings_full = ratings.fillna(0) #fill the NaN with the mean of each column
ratings_full.head(3)
```

```python
#finding correlations using corrwith

correlation = ratings_full.corrwith(ratings_full['Forrest Gump (1994)'], axis=0, drop=False)
correlation.head(5)  #first five results in the data frame
```

```python
correlation.sort_values(ascending = False).head(10)  # the top ten most correlated movies in the dataframe
```
