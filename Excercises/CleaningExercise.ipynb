```python
#import relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load csv data and plot in data frame
df1 = pd.read_csv("Assignment 1 - survey data.csv")
df1.head()
```

```python
#load data with semi-colon seperator
df2 = pd.read_csv("Assignment 1 - steps data.csv", sep=';')
df2.head()
```

```python
#combine the two data frame
df = pd.merge(df1, df2, on = 'id')
df.head(10)
```

```python
df.info()
```

```python
df.waist.value_counts()
```

```python
#plot histogram with 10 bins to get distribution of waist measurement
waist = df.sort_values("waist", ascending = True)
df.plot(kind='hist', y = "waist", bins = 10)
```

```python
df['waist'] = df['waist'].apply(pd.to_numeric, errors='coerce')
#To remove the values above 5 we need to create a function that we then apply to the entire column
def below_25(x):
    if(x < 25.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(below_25)
```

```python
#check if values below 25 removed
waist = df.sort_values("waist", ascending = True)
df.plot(kind='hist', y = "waist", bins = 10)
```
